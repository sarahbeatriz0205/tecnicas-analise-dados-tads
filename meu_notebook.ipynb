### Ordenação com sort_values()

```python
df.sort_values(by=['release_year'], ascending=True)
```

### Agrupamento com groupby()

```python
df.groupby('genre')['critic_score'].sum() # somando os scores agrupando pelo gênero
df.groupby('genre')['critic_score'].min() # retornando os scores mínimos desse agrupamento
```

- **reset_index():** Cria um novo dataframe baseado no agrupamento e no método usado

```python
df.groupby('genre')['critic_score'].sum().reset_index() # cria um novo dataframe baseado no agrupamento e no método usado
```
