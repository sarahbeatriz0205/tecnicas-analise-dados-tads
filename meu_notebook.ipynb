### Gráficos

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.countplot(data=df, y='genre', order=df['genre'].value_counts().index)
plt.title('Distribuição da Variável Descrição')
plt.xlabel('Contagem')
plt.ylabel('Gênero')
plt.show()
```

- interativo com plotly

```python
import plotly.express as px

counts_data = df['genre'].value_counts().reset_index()
counts_data.columns = ['genre', 'Contagem']

grafico = px.bar(
    counts_data,
    x='Contagem',
    y='genre',
    title="Contagem de ocorrências dos gêneros dos jogos mais vendidos",
    category_orders={'genre': counts_data['genre'].tolist()},
    text='Contagem'
)

grafico.update_traces(textposition='outside')
grafico.show()
```

### Análise de variáveis quantitativas

```python
critic_score = df["critic_score"].describe()
critic_score
```

### Histograma e Boxplot de uma das variáveis quantitativas do dataset

> Boxplot

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='critic_score')
plt.axvline(critic_score['50%'], color='red', linestyle='--', label=f'Mediana: {critic_score["50%"]:.2f}')
plt.title('Box Plot da Variável Nota da Crítica (critic_score) com Mediana Destacada')
plt.xlabel('Nota da crítica')
plt.legend()
plt.show()
```

> Histograma

```python
import pandas as pd
import kagglehub
import matplotlib.pyplot as plt
import seaborn as sns

# Carregar o DataFrame df, caso não esteja carregado
path = kagglehub.dataset_download("bhushandivekar/video-game-sales-and-industry-data-1980-2024") # aqui ele baixa uma pasta com os dados
df = pd.read_csv(path + "/Video_Games_Sales_Cleaned.csv")
bin_size=len(df["critic_score"].value_counts()) # Total de valores difeentes
plt.figure(figsize=(10, 6))
ax = sns.histplot(data=df, x='critic_score', bins='auto') # Changed stat to 'count' and adjusted bins
plt.title('Distribuição da Variável Nota da Crítica com Contagens')
plt.xlabel('critic_score')
plt.ylabel('Contagem') # Rótulo alterado para Contagem

# Add labels on top of bars
for p in ax.patches:
    # Get the height of the bar. It's the count.
    height = p.get_height()
    # If the height is 0, skip it to avoid clutter
    if height > 0:
        ax.text(p.get_x() + p.get_width() / 2., # x-coordinate of text (center of bar)
                height, # y-coordinate of text (top of bar)
                f'{int(height)}', # text to display (integer count)
                ha='center', va='bottom', fontsize=8) # horizontal and vertical alignment
plt.show()
```

```python
import plotly.express as px
grafico=px.histogram(df, x='critic_score', title='Distribuição da Variável Nota da Crítica',labels={'critic_score':'critic_score'},marginal="violin")
grafico.show()
```
